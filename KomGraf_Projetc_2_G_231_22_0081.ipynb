```python
plt.figure(figsize=(15,5))

# Menentukan posisi bulan
bulan = np.arange(1,13)

# Membuat bar chart
plt.bar(bulan, df['1981 - 1990'], width = 0.2, color='royalblue', label='1981 - 1990')
plt.bar(bulan + 0.2, df['1991 - 2000'], width = 0.2, color='firebrick', label='1991 - 2000')
plt.bar(bulan + 0.2 * 2, df['2001 - 2010'], width = 0.2, color='yellowgreen', label='2001 - 2010')

# Mengatur label sumbu x dan y, serta judul
plt.xlabel('Bulan', weight='bold')
plt.ylabel('Curah Hujan (mm)', weight='bold')

# Mengatur posisi dan label pada sumbu x
plt.xticks(bulan + 0.2, bulan)

# Menambahkan garis-garis horizontal kecil di belakang
for i in range(20, 700, 20):
    plt.axhline(y=i, color='gray', linestyle='-', alpha=0.3)

# Menampilkan legenda
plt.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.2))

# Menampilkan plot
plt.show()
```

```python
data1 = [23, 25.2, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.9]
data2 = [23.4, 25.3, 25.1, 25.5, 24.9, 25, 24.5, 24, 26]
data3 = [24, 25, 23.8, 23.9, 24, 24.1, 24, 22.9, 23]
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

plt.figure(figsize=(13, 5))

plt.plot(dates, data1, marker='o', color='royalblue', label='Asimilasi')
plt.plot(dates, data3, marker='o', color='gold', label='Observasi')
plt.plot(dates, data2, marker='o', color='red', label='Non Asimilasi')

plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')
plt.legend(loc="lower right")

for d in [20,21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())


plt.show()
```
